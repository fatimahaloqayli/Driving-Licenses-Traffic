````python
# creating a dictionary for traffic accidents dataset
traffic_Acci_dtype_lst = ['accidents' for i in range (len(accidents.columns))]

traffic_descrip_lst =[ 'The region where the triffic accidents exist', 'The year where the triffic accidents exist','The x corrdinate where the traffic accidents, deads, injuries exist',
                    'The y corrdinate where the traffic accidents', 'The number of accidents exist per year/regoin', 'The number of deads  per year/regoin', 'The number of injured people  per year/regoin']
                    
traffic_dict = {'Feature': accidents.columns , 'Type': list(accidents.dtypes), 'Dataset': traffic_Acci_dtype_lst, 'Description': traffic_descrip_lst}

#convert the dictionary into pandas dataset
traffic_dict=pd.DataFrame.from_dict(traffic_dict)
````

#### Statistical Evaluation of Distributions 

**If you feel it's appropriate**, using methods we discussed in class, run hypothesis tests to compare variables of interest in our dataset. 

# Code

I uesd T-test  in order to determine whether there is statistical evidence that the associated population means are significantly different.('driving_licenses'and 'num_of_accidents')

````python
from scipy.stats import ttest_ind
import numpy as np
driving_licenses = result1['driving_licenses']   
 
num_of_accidents = result1['num_of_accidents']
print(driving_licenses)
print("num_of_accidents data :-\n")
print(num_of_accidents)
driving_licenses_mean = np.mean(driving_licenses)
num_of_accidents_mean = np.mean(num_of_accidents)
print("driving_licenses mean value:",driving_licenses_mean)
print("num_of_accidents mean value:",num_of_accidents_mean)
driving_licenses_std = np.std(driving_licenses)
num_of_accidents_std = np.std(num_of_accidents)
print("driving_licenses std value:",driving_licenses_std)
print("num_of_accidents std value:",num_of_accidents_std)
ttest,pval = ttest_ind(driving_licenses,num_of_accidents)
print("p-value",pval)
if pval <0.05:
  print("we reject null hypothesis")
else:
  print("we accept null hypothesis")
````
